```python
import pandas as pd
import json
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

voting_data = pd.read_csv("https://www.electoralcommission.org.uk/__data/assets/file/0014/212135/EU-referendum-result-data.csv")
county_vote_df = voting_data[['Region','Area', 'Electorate', 'Pct_Turnout', 'Valid_Votes', 'Remain', 'Leave']]
county_vote_df.name = 'county_vote_df'
region_vote_df = voting_data[['Region', 'Electorate', 'Pct_Turnout', 'Valid_Votes', 'Remain', 'Leave']].groupby('Region').agg(np.sum)
region_vote_df.name = 'region_vote_df'

dataframes = [county_vote_df, region_vote_df]
for dataframe in dataframes:
    dataframe.loc[:,'Pct_Remain'] = dataframe['Remain']/dataframe['Valid_Votes']
    dataframe.loc[:,'Pct_Leave'] = dataframe['Leave']/dataframe['Valid_Votes']
     
    output_file = os.path.join('..', 'Resources', dataframe.name +'.csv')
    #  Open the output file
    with open(output_file, 'w') as datafile:
        dataframe.to_csv(output_file)
        
strong_leave_df = county_vote_df[county_vote_df["Pct_Leave"] > .65]
strong_leave_df.name = 'strong_leave_df'
output_file = os.path.join('..', 'Resources', strong_leave_df.name +'.csv')
with open(output_file, 'w') as datafile:
        strong_leave_df.to_csv(output_file)
strong_remain_df = county_vote_df[county_vote_df["Pct_Remain"] > .65]
strong_remain_df.name = 'strong_remain_df'
output_file = os.path.join('..', 'Resources', strong_remain_df.name +'.csv')
with open(output_file, 'w') as datafile:
        strong_remain_df.to_csv(output_file)
```
